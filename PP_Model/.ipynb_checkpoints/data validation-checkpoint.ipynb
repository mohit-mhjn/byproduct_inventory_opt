```python
import pandas as pd
```

```python
fresh_sold = pd.read_csv('Fresh_Product_Sold.csv')
fresh_sold
```

```python
fresh_sold1 = pd.DataFrame(fresh_sold.days.str.split('_',1).tolist(),
                                   columns = ['Products','Customer'])
```

```python
Fresh_sold = pd.concat([fresh_sold, fresh_sold1], axis=1)
```

```python
Fresh_sold
```

```python
df = Fresh_sold
del df['days']
```

```python
df_1 = df[(df['Customer']=='1')]
```

```python
df_2 = df[df['Customer']=='2']
```

```python
df_final = pd.concat([df_1.T,df_2.T],axis=1)
```

```python
df_final
```

```python
def sjoin(x): return ';'.join(x[x.notnull()].astype(str))
df_final.groupby(level=0, axis=1).apply(lambda x: x.apply(sjoin, axis=1))
```

```python
df.pivot_table(values=['Ribs','Keels','Legs','Thigh','Wings','Breasts','Drumsticks'],index=['Day_1','Day_2','Day_3','Day_4'])
```

```python
df.set_index(['time'], inplace=True)

# iterate through the list of countries, and create a plot
for customer in df.Customer.unique():
    # .plot.bar makes bar charts
    ax = df[df.Country==country].plot.bar(stacked=True)
    # make room for the legend, and sets in the upper right corner
    ax.set_xlim(ax.get_xlim()[0],ax.get_xlim()[1]+1)
    ax.legend(loc='upper right')
    # add the country name to the top of the chart.
    ax.set_title(country)
```
